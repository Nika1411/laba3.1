Узнали самое длинное название города в таблице

```python
%%time
list=[]
for c in city["city"]:
    c = str(c)
    list.append(c)
print("Длина: ", len(max(list, key=len)), "\nСлово: ", max(list, key=len))
```

Написали запрос, который посчитал кол-во городов для каждого часового пояса в Сибирском и Приволжском федеральных округах. Значение timezone(UTC +5) для city_count = 58

```python
k = city[city['federal_district'].isin(['Приволжский', 'Сибирский'])]
k[['timezone','country']].groupby('timezone').count().rename({'country': 'city_count'}, axis=1)
```

Решили задачу: написали  запрос, который посчитал кол-во городов в каждом часовом поясе. Отсортировали по кол-ву городов по убыванию.

```python
k = city[['timezone','address']].groupby('timezone').count().sort_values('address', ascending=False)
k.rename({'address': 'city_count'}, axis=1)
```

Выбрали свою таблицу, и произвели те же запросы, что и в sqlite
